```python
def barchart(inPersonDF,mixedDF,onlineDF,demographic,state,candidate):
    '''
    The purpose of this function is to take three existing DataFrames 
    and pull the polling data from those DataFrames to create a new
    DataFrame for a user-given demographic and candidate
    
    :inPersonDF: a DataFrame that has the polling averages for all the in-person polls 
    :mixedDF: a DataFrame that has the polling data for the part online, part in-person polls
    :onlineDF: a DataFrame that has the polling data for the completely online polls
    :demographic: this is the demographic that we want the side-by-side polling data for
    :candidate: this is the candidate that we want the above information for
    :state: this is the state for which we want the demographic data
    '''
    assert isinstance(inPersonDF,pd.DataFrame)
    assert isinstance(mixedDF,pd.DataFrame)
    assert isinstance(onlineDF,pd.DataFrame)
    assert isinstance(state,str)
    assert isinstance(demographic,str)
    assert isinstance(candidate,str)
    #columns = ["In-Person","Mixed","Online"]
    #demographic = str(demographic) #convert to a string if it isn't already a string, assuming it's a string already
    index = [demographic] 
    
    preElectionInPerson = [float(0)]#average of all A+ pollsters
    preElectionMix = [0] #A- pollster that had part online portion
    preElectionPureOnline = [0] #SurveyMonkey
    df = pd.DataFrame({'In-Person': preElectionInPerson, 
                    'Mixed': preElectionMix,
                    'Online': preElectionPureOnline},
                    index=index)
    if (candidate == "Trump"):
        inPersonDF = inPersonDF[~inPersonDF['Candidate'].isin(['Biden'])]
        mixedDF = mixedDF[~mixedDF['Candidate'].isin(['Biden'])]
        onlineDF = onlineDF[~onlineDF['Candidate'].isin(['Biden'])]
    if (candidate == "Biden")
        inPersonDF = inPersonDF[~inPersonDF['Candidate'].isin(['Trump'])]
        mixedDF = mixedDF[~mixedDF['Candidate'].isin(['Trump'])]
        onlineDF = onlineDF[~onlineDF['Candidate'].isin(['Trump'])]
        
    inPersonDF = inPersonDF[inPersonDF["State"].str.contains(state)]
    mixedDF = mixedDF[mixedDF["State"].str.contains(state)]
    onlineDF = onlineDF[onlineDF["State"].str.contains(state)]
    
    df['In-Person'] = inPersonDF[demographic]
    df['Mixed'] = mixedDF[demographic]
    df['Online'] = onlineDF[demographic]
```
