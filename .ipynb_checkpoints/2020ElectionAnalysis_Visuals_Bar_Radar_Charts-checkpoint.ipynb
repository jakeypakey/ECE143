```python
import processing 
import visuals_bar_radar_chart
%matplotlib inline
```

```python
#phone polling was conducted by Marist, Monmouth, and Siena, so we want the average of the three
phoneDF = processing.avgPolls(processing.readPreElectionPolls(),['Marist','Monmouth','Siena'])
#mixed polling was only done by Emerson, but calling avgPolls is okay because the average of Emerson is itself
mixedDF = processing.avgPolls(processing.readPreElectionPolls(),['Emerson'])
#online polling was done by SurveyMonkey, and calling avgPolls is okay because the average of SurveyMonkey is itself
onlineDF = processing.avgPolls(processing.readPreElectionPolls(),['SurveyMonkey'])
```

```python
#for reference, let's see what the dataframes look like before they are processed for the bar chart
phoneDF.head()
```

```python
mixedDF.head()
```

```python
onlineDF.head()
```

```python
#the following is a demonstration of the barchart processing function whose output will serve as the input
#to the plot bar chart function
df = visuals_bar_radar_chart.barchart_processing(phoneDF,mixedDF,onlineDF,"Independent","Florida","Trump")
```

```python
#let's see what the dataframe looks like now that it's been processed in a way that can be used for plotting 
#the bar chart
df.head()
```

```python
#now that we have our dataframe, let's use it to plot a bar chart!
visuals_bar_radar_chart.plot_bar_chart(df,'Independent','Florida','Trump')
```

```python
#in the event we want to only plot the phone and mixed polling data, we will do the processing 
#using the other function, barchart_processing2
df = visuals_bar_radar_chart.barchart_processing2(phoneDF,mixedDF,"Independent","Florida","Trump")
```

```python
#let's see how the processed dataframe looks now compared to before
df.head()
#notice that there are now 2 values rather than 3
```

```python
#plot the bar chart based on the two polling methodologies
visuals_bar_radar_chart.plot_bar_chart(df,'Independent','Florida','Trump')
```

```python
#notice how 2 bars are now plotted instead of 3
```
